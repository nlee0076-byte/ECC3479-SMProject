```python
# NUMERIC + NUMERIC SCATTER PLOT
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("data/clean/cleaned_social_media.csv")


plt.figure(figsize=(10, 6))
sns.scatterplot(
   data=df,
   x="avg_daily_usage_hours",
   y="mental_health_score",
   hue="gender",
   style="usage_group",
   palette="deep",
   alpha=0.8,
   s=80
)
plt.title("Average Daily Social Media Usage vs Mental Health Score")
plt.xlabel("Average Daily Usage (hours)")
plt.ylabel("Mental Health Score")
plt.legend(title="Gender / Usage group", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()

```

```python
# NUMERIC + CATEGORICAL BOX PLOT
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("data/clean/cleaned_social_media.csv")


plt.figure(figsize=(12, 7))
sns.boxplot(
   data=df,
   x="usage_group",
   y="mental_health_score",
   hue="gender",
   palette="pastel"
)


plt.title("Mental Health Score Distribution by Usage Group and Gender")
plt.xlabel("Usage Group")
plt.ylabel("Mental Health Score")
plt.legend(title="Gender", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
# CATEGORICAL + CATEGORICAL COUNT PLOT
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("data/clean/cleaned_social_media.csv")


# 1) Countplot with hue
plt.figure(figsize=(10, 6))
sns.countplot(
   data=df,
   x="usage_group",
   hue="gender",
   palette="pastel"
)
plt.title("Count of Usage Group by Gender")
plt.xlabel("Usage Group")
plt.ylabel("Count")
plt.legend(title="Gender")
plt.tight_layout()
plt.show()


# 2) Crosstab table
ct = pd.crosstab(df["usage_group"], df["gender"])
print(ct)


# 3) Heatmap of counts
plt.figure(figsize=(8, 5))
sns.heatmap(
   ct,
   annot=True,
   fmt="d",
   cmap="YlGnBu",
   cbar=True
)
plt.title("Usage Group vs Gender Count Heatmap")
```

```python
# Faceting the same numeric relationship across usage groups
sns.set_style("whitegrid")
g = sns.relplot(
data=df,
x="avg_daily_usage_hours",
y="mental_health_score",
col="usage_group",
hue="gender",
kind="scatter",
col_wrap=3,
palette="deep",
height=4,
aspect=1
)

g.set_axis_labels("Average Daily Usage (hours)", "Mental Health Score")
g.set_titles("{col_name} usage group")
g.add_legend(title="Gender")
plt.tight_layout()
plt.show()
```
